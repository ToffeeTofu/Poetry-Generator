```python
# From transformers - Hugging Face library for pretrained language models
# AutoTokenizer - converts text to tokens and tokents to text
# AutoModelForCasualLM - Loads casual language model - predicts the next token given previous ones
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
# torch - for tensors
import torch

# Load model
print("Loading AI model...")
# DistilGPT - smaller + faster GPT2 --> more efficient
model_name = "gpt2"
# tokenizer --> knows GPT2's vocab and tokenization rules
# tokenizer and model must match or predictions will break since words will be
# tokenzied incorrectly
tokenizer = AutoTokenizer.from_pretrained(model_name)
# loads the neural network with pretrained weights
model = AutoModelForCausalLM.from_pretrained(model_name)
import random
```

```python
# Simple example
text = "FROM off a hill whose concave"

# --- Poetry structure ---
num_lines = 4
line_lengths = [random.randint(6, 9) for _ in range(num_lines)]
temperature = 0.8
poem_lines = []

context = text  # keeps model memory flowing
poem_lines.append(text)

# Generate 4 lines
for line_index in range(num_lines):

    line_text = ""
    words = 0
    # Keep going as long as number of words is less then line_length
    # Generates a new line_length for every line between 6 and 9
    while words < line_lengths[line_index]:
        input_ids = tokenizer.encode(context, return_tensors="pt")

        with torch.no_grad():
            outputs = model(input_ids)
            predictions = outputs.logits

        logits = predictions[0, -1, :]
        probs = torch.softmax(logits / temperature, dim=0)

        top_probs, top_indices = torch.topk(probs, 7)
        choice = random.randrange(7)

        next_word = tokenizer.decode([top_indices[choice]])

        line_text += next_word
        context += next_word

        # Count only visible words
        words = len(line_text.strip().split())

    # Adds the new line to the poem_lines array
    poem_lines.append(line_text.strip())

print("\nFinal poem:\n")
# Joins all the lines together
print("\n".join(poem_lines))
```
